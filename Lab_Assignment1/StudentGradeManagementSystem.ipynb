```python
# In the space below, please import the Pandas library
# (1 mark)

import pandas as pd;
```

```python
# Load the student grade CSV files into DataFrames using the defined variables
# (2 marks)
df_english = pd.read_csv('studentGrades_English.csv')
df_math = pd.read_csv('studentGrades_Math.csv')
```

```python
# Merging the DataFrames on 'studentID'
df_studentGrades = pd.merge(df_english, df_math, on='studentID', suffixes=('_English', '_Math'))
print(df_studentGrades)
```

```python
# Take a look at the data contained in the csv files. Why is studentID chosen as the key?
# (1 mark)
# Answer is written in the textbox below
```

**Answer**
The studentID is chosen as the key because it is a unique identifier for each student across the datasets. Using studentID ensures that the correct grades from both English and Math are associated with the right student, as it avoids ambiguity that could arise if other fields (like name) were used, which may not be unique. This maintains data integrity and allows for accurate merging of records.

In this case, .iloc makes more sense because it allows us to access rows based on their positional index, which is useful when we want to get the first row (the student with the highest GPA) from df_studentsGPASorted after sorting.



1.   .iloc is for position-based indexing. After sorting, the highest GPA is guaranteed to be in the first position (index 0), so .iloc[0] will always retrieve it.
2.   .loc is for label-based indexing, meaning it would retrieve rows based on the index labels (which may not change when sorting), so it's less reliable for accessing sorted data.

Heres a breakdown of what each line in the code is doing:

1) **df_studentsGPASorted_homeroom['studentRank'] = df_studentsGPASorted_homeroom.groupby('homeRoom').cumcount() + 1:**



1.   This line creates a new column called studentRank.
2.   The groupby('homeRoom').cumcount() function assigns a sequential number to each student within their respective homeroom group based on the sorting order (GPA in this case).
3.   Adding +1 makes the ranks start from 1 (instead of 0), assigning each student a rank within their homeroom based on GPA.



2) **df_studentsGPASorted_homeroom['newIndex'] = df_studentsGPASorted_homeroom['homeRoom'].astype(str) + '-' + df_studentsGPASorted_homeroom['studentRank'].astype(str):**



1.   This line creates a new column called newIndex by concatenating the homeRoom and studentRank columns.
2.   The astype(str) ensures both homeRoom and studentRank are treated as strings, so they can be combined using + to form a new index (e.g., A-1, B-2).
3.   This makes the new index more meaningful by showing both homeroom and rank in the index.


3) **df_studentsGPASorted_homeroom = df_studentsGPASorted_homeroom.set_index('newIndex'):**



1.   This line sets the newIndex column (created above) as the new index for the DataFrame.
2.   The previous default numerical indices are replaced with the new, more intuitive index, which now includes the homeroom and the students rank.


4) **df_studentsGPASorted_homeroom = df_studentsGPASorted_homeroom.drop(columns=['studentRank']):**



1.   After setting the new index, the studentRank column is no longer
needed, so this line drops it from the DataFrame.
2.   The DataFrame is now cleaner, with just the relevant columns and the new meaningful index.
